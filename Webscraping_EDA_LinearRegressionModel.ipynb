```
# Data set

# X = df.drop(['Player','Pos','Tm','Rk','PTS'], axis=1)

# X = df[["2P","3P","G","GS","MP","FG"]]

X = df[["G","3P"]]
Y = df.iloc[:,-1]

# Convert to float
X = X.astype(float, errors = 'raise')
Y = Y.astype(float, errors = 'raise')

X
```

### Model from original X-Y pairs

```
# Model from original X-Y pairs
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Luciano code
from sklearn.metrics import accuracy_score

# Data set
# X = dataset.drop(['logS'], axis=1)
# Y = dataset.iloc[:,-1]

# Data split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

# Y-Shuffling


# Model building
model = LinearRegression()
model.fit(X_train, Y_train)





# previsoes = modelo.predict(teste_x)

# acuracia = accuracy_score(teste_y, previsoes) * 100
# print("A acurácia foi %.2f%%" % acuracia)





# Make prediction - Train
Y_train_pred = model.predict(X_train)

# Make prediction - Test
Y_test_pred = model.predict(X_test)




# Compute R2 score - Train
r2a = r2_score(Y_train, Y_train_pred)

# Compute R2 score - Test
r2a_t = r2_score(Y_test, Y_test_pred)




# Print R2 score - Train
print('Train R² = %s' % r2a.round(2))

# Print R2 score - Test
print('Test R² = %s' % r2a_t.round(2))


# asd = accuracy_score(Y_train, Y_train_pred)
# print('Accuracy = %s' % ac_score.round(2))
```

```
# Plot of predicted vs actual
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5,5))
plt.scatter(x=Y_train, y=Y_train_pred, c="#7CAE00", alpha=0.3)

# plt.scatter(x=Y_test, y=Y_test_pred, c="#7CAE00", alpha=0.3)

# Add trendline
# https://stackoverflow.com/questions/26447191/how-to-add-trendline-in-python-matplotlib-dot-scatter-graphs
z = np.polyfit(Y_train, Y_train_pred, 1)
p = np.poly1d(z)

plt.plot(Y,p(Y),"#F8766D")
plt.ylabel('Predicted Points')
plt.xlabel('Points')
```
